Ciągi znaków cz1 - podstawy

```python
print('Ciągi znaków cz1 - podstawy')
print('\n\n')

zmienna_int1 = 2
zmienna_int2 = 3
zmienna_float = 2.3
zmienna_txt = "abcdefghijklmnoprstuwyz" 

print('wartość zmienna_int:\t', zmienna_int1, ', typ zmienna_int1: ', type(zmienna_int1))
print('wartość zmienna_int:\t', zmienna_int2, ', typ zmienna_int2: ', type(zmienna_int2))
print('wartość zmienna_float:\t', zmienna_float, ', typ zmienna_float: ', type(zmienna_float))
print('wartość zmienna_txt: ', zmienna_txt, ', typ zmienna_txt: ', type(zmienna_txt))

print('\n')

suma1 = zmienna_float + zmienna_int1
print('wartość suma1 = zmienna_float + zmienna_int1: ', suma1, '=', zmienna_float, '+', zmienna_int1, ', typ suma1: ', type(suma1))
suma1 = zmienna_float * zmienna_int1
print('wartość suma1 = zmienna_float * zmienna_int1: ', suma1, '=', zmienna_float, '*', zmienna_int1, ', typ suma1: ', type(suma1))
suma1 = zmienna_float ** zmienna_int1
print('wartość suma1 = zmienna_float ** zmienna_int1: ', suma1, '=', zmienna_float, '**', zmienna_int1, ', typ suma1: ', type(suma1))
suma1 = zmienna_int1 / zmienna_int2
print('wartość suma1 = zmienna_int1 / zmienna_int2: ', suma1, '=', zmienna_int1, '/', zmienna_int2, ', typ suma1: ', type(suma1))
suma1 = zmienna_int1 % zmienna_int2
print('wartość suma1 = zmienna_int1 % zmienna_int2: ', suma1, '=', zmienna_int1, '%', zmienna_int2, ', typ suma1: ', type(suma1))

print('\n')

print('zmienna_txt: ', zmienna_txt)
print('zmienna_txt[0] ', zmienna_txt[0], ', zmienna_txt[2]', zmienna_txt[2], ', zmienna_txt[4]', zmienna_txt[4])
print('zmienna_txt[0] ', zmienna_txt[0], ', zmienna_txt[-1]', zmienna_txt[-1], ', zmienna_txt[-3]', zmienna_txt[-3])
print('zmienna_txt[0:10:2] ', zmienna_txt[0:10:2])
print('zmienna_txt[-1:-10:2] ', zmienna_txt[-1:-10:-2])
print('zmienna_txt[5:] ', zmienna_txt[5:])
print('zmienna_txt[-5:] ', zmienna_txt[-5:])
print('zmienna_txt[:10] ', zmienna_txt[:10])
print('zmienna_txt[:-10] ', zmienna_txt[:-10])
print('zmienna_txt[::] ', zmienna_txt[::])

print('\n')

print('długość len(zmienna_txt) \t', len(zmienna_txt))
```

Ciągi znaków cz2 - metody

```python
print('Ciągi znaków cz2 - metody')
print('\n\n')

z_imie = "Daniel"
z_nazwisko = "Mierzejewski"

print('z_imie i z_nazwisko ',z_imie, ' ',z_nazwisko)
print('len(z_imie) ', len(z_imie))
print('len(z_nazwisko) ', len(z_nazwisko))

z_inicjaly = z_imie[0] + z_nazwisko[0]
print('z_inicjaly ', z_inicjaly)

z_inicjaly = z_inicjaly + " to moje inicjały"
print(z_inicjaly)

print('\n')

tmp1 = z_imie.upper() + ' ' + z_nazwisko.upper()
print('funkcja upper() ', tmp1)
tmp1 = z_imie.lower() + ' ' + z_nazwisko.lower()
print('funkcja lower() ', tmp1)

print('\n')

tmp2 = z_imie + ' ' + z_nazwisko
print('tmp2 przed podziałem ', tmp2)
print('type(tmp2) ', type(tmp2))
print('tmp2[0] ', tmp2[0])
print('len(tmp2) ', len(tmp2))
tmp2 = tmp2.split()
print('tmp2.split() po podziale ', tmp2)
print('type(tmp2) ', type(tmp2))
print('tmp2[0] ', tmp2[0])
print('len(tmp2) ', len(tmp2))
print('\n')
tmp2 = z_imie + ' ' + z_nazwisko
print('tmp2 przed podziałem ', tmp2)
print('type(tmp2) ', type(tmp2))
print('tmp2[0] ', tmp2[0])
print('len(tmp2) ', len(tmp2))
tmp2 = tmp2.split('e')
print("tmp2.split('e') po podziale ", tmp2)
print('type(tmp2) ', type(tmp2))
print('tmp2[0] ', tmp2[0])
print('len(tmp2) ', len(tmp2))

```
