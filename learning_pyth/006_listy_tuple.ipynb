```python
print('Listy cz1 - podstawy')
print('\n\n')

lista_int1 = [5,6,3,2,4,7,8,1,9]
lista_float1 = [3.14,8.45,3.5]
lista_str1 = ['Ala', 'ma', 'kota', 'a', 'kot', 'ma', 'Alę', 'gdzieś']
lista_mieszana1 = ['foo', 5, 3.25, 'bar']
tupla_mieszana1 = {1.5, 'dwa tory', 2}

print('type(lista_int1)\t',type(lista_int1))
print('type(lista_float1)\t',type(lista_float1))
print('type(lista_str1)\t',type(lista_str1))
print('type(lista_mieszana1)\t',type(lista_mieszana1))
print('type(tupla_mieszana1)\t',type(tupla_mieszana1))

print('\n')

print('a w skrócie tupla to lista, której elementów i wielkości nie można zmienić')
print('a wartości list w przeciwieństwie do łańcuchów można zmieniać\n')

print('Długość lista_int1: ', len(lista_int1))
print('Długość lista_float1: ', len(lista_float1))
print('Długość lista_str1: ', len(lista_str1))
print('Długość lista_mieszana1: ', len(lista_mieszana1))
print('Długość tupla_mieszana1: ', len(tupla_mieszana1))
print('piknie działa funkcja len() :D')

print('\n')

print('lista_str1[2:6]:\t', lista_str1[2:6])
print('lista_str1[::2]:\t', lista_str1[::2])
print('lista_str1[::2]:\t', lista_str1[::2])
print('lista_str1[-1:-6:-2]:\t', lista_str1[-1:-6:-2])

print('Fajne, co? Lecimy dalej z koksem... \n')

print('Działania na listach? Spróbujmy')
suma1 = lista_int1 + lista_float1
print('suma1 = lista_int1 + lista_float1: ',lista_int1,' + ',lista_float1,' = ',suma1)
suma2 = lista_int1[3] + lista_float1[2]
print('suma2 = listaint1[3] + lista_float1[2]: ',lista_int1[3],' + ',lista_float1[2],' = ',suma2)
suma3 = lista_int1[0] * lista_float1[1]
print('suma3 = lista_int1[0] * lista_float1[1]: ',lista_int1[0],' * ',lista_float1[1],' = ',suma3)
suma4 = lista_str1[3] * lista_mieszana1[1]
print('suma4 = lista_str1[3] * lista_mieszana1[1]: ',lista_str1[3],' * ',lista_mieszana1[1],' = ',suma4)

print('\n')

print('Inne zabawy na listach')
print('Stara lista_str1:\t',lista_str1)
lista_str1.append('!')
print('Nowa lista_str1:\t',lista_str1)
print('Usuwanie ostatniej wartości z lista_str1: ',lista_str1.pop())
print('Nowa lista_str1:\t',lista_str1)
print('Usuwanie pierwszej wartości z lista_str1: ',lista_str1.pop(0))
print('Nowa lista_str1:\t',lista_str1)
print('Sortowanie od najniższej lista_str1:\t',lista_str1.sort())
print('Nowa lista_str1:\t',lista_str1)
print('Sortowanie od najwyższej lista_str1:\t',lista_str1.reverse())
print('Nowa lista_str1:\t',lista_str1)
lista_str1[0] = 'nie ma'
print('Nowa lista_str1:\t',lista_str1)
```
